```
from newsapi import NewsApiClient

newsapi = NewsApiClient(api_key='f86d016024764c88a4db5f61940a1cae')
```

```
all_articles = newsapi.get_everything(domains='abs-cbn.com',
                                      from_param='2021-03-11',
                                      to='2021-03-12',
                                      language='en',
                                      page_size=100)

print(all_articles)
```

```
articles_json = []

for articles in all_articles.get('articles'):

  date=articles["publishedAt"]
  title=articles["title"]
  content=articles["content"]
  author=articles["author"]
  articles_json.append({
            "date": date,
            "title": title,
            "content": content,
            "author": author
        })

print(articles_json)
```

```
from google.colab import files
import json

x = json.dumps(articles_json)
with open('ABS-CBN_Articles.json', 'w') as f:
  f.write(str(x))

files.download('ABS-CBN_Articles.json')
```
