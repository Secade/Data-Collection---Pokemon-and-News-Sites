```
import requests
URL = "https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"

page = requests.get(URL)
```

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```
poke_content = soup.find(id='mw-content-text')
poke_tables = poke_content.find_all('table')
```

```
gen2_list = poke_tables[2]
gen3_list = poke_tables[3]
gen4_list = poke_tables[4]
gen5_list = poke_tables[5]
gen6_list = poke_tables[6]
gen7_list = poke_tables[7]
gen8_list = poke_tables[8]
```

```
gen2_list.contents
```

```
def getPokemonPerGeneration(genList):
  info_start=3

  for i in range(info_start, len(genList.contents), 2):
    poke_info=genList.contents[i]
    dex=poke_info.contents[1].text.strip()
    ndex=poke_info.contents[3].text.strip()
    name=poke_info.contents[7].text.strip()
    page=f'https://bulbapedia.bulbagarden.net/wiki/{name}_(Pok%C3%A9mon)'
    type1=poke_info.contents[9].text.strip()
    if len(poke_info.contents) > 10:
        type2=poke_info.contents[11].text.strip()
        print(f'Pokemon {ndex} ({dex}) {name} is a {type1} & {type2} Pokemon. Pokemon Page: {page}')
    else:
        print(f'Pokemon {ndex} ({dex}) {name} is a {type1} Pokemon. Pokemon Page: {page}')
```

```
getPokemonPerGeneration(gen2_list)
```

```
from google.colab import files
import json
```

```
def saveGenToJSON(genList,kdex,fileName):
  info_start=3
  gen_json = []

  for i in range(info_start, len(genList.contents), 2):
    poke_info=genList.contents[i]
    dex=poke_info.contents[1].text.strip()
    ndex=poke_info.contents[3].text.strip()
    name=poke_info.contents[7].text.strip()
    page=f'https://bulbapedia.bulbagarden.net/wiki/{name}_(Pok%C3%A9mon)'
    type1=poke_info.contents[9].text.strip()
    if len(poke_info.contents) > 10:
        type2=poke_info.contents[11].text.strip()
        gen_json.append({
            f"{kdex}": dex,
            "ndex": ndex,
            "name": name,
            "type1": type1,
            "type2": type2,
            "page": page
        })
    else:
        gen_json.append({
            f"{kdex}": dex,
            "ndex": ndex,
            "name": name,
            "type1": type1,
            "page": page
        })
  print(gen_json)

  x = json.dumps(gen_json)
  with open(f'{fileName}.json', 'w') as f:
    f.write(str(x))

  files.download(f'{fileName}.json')
```

```
saveGenToJSON(gen2_list,"jdex","Gen2List")
```
